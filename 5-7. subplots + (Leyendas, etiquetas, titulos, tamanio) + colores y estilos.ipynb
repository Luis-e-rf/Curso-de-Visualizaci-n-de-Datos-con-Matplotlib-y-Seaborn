```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,1].plot(x,np.sin(x),'r')
axes[1,0].plot(x,np.tan(x),'y')
axes[0,3].plot(x,np.sin(x)**2,'g')
fig.tight_layout()
```

```python
fig, axes = plt.subplots(1,2,figsize=(8,5))
axes[0].plot(x,np.cos(x),'b',label='$cos(x)$')
axes[0].set_title('Relacion de x - cos(x)')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(x,np.sin(y),'r')
axes[1].set_title('Relacion de x - sin(y)')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
fig.tight_layout()
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es el titulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left',bbox_to_anchor=(1,0.9))
```

# Estilo Pyplot

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,x+1,color = 'red', linewidth=5, linestyle='--', alpha=0.5, marker='P', markerfacecolor='#000000')
ax.plot(x,x+2,color = 'green', linewidth=3, linestyle='dashed', marker='X')
ax.plot(x,x+3,color = 'blue', linewidth=8, linestyle=':', marker='o')
ax.plot(x,x+4,color = '#000000', linewidth=1, linestyle='-', marker='v', markersize=20)
```

```python
# Crear la figura principal y los subplots
fig, ax = plt.subplots(4, 7, figsize=(15, 15))
contador = 0

# Iterar sobre los ejes y aplicar los estilos temporalmente
for i in range(4):
    for j in range(7):
        if contador < len(plt.style.available):
            # Crear una figura temporal para aplicar el estilo
            with plt.style.context(plt.style.available[contador]):
                temp_fig, temp_ax = plt.subplots()
                temp_ax.plot(x, y)
                
                # Renderizar el gráfico temporal en una imagen usando buffer_rgba
                temp_fig.canvas.draw()
                image = np.frombuffer(temp_fig.canvas.buffer_rgba(), dtype=np.uint8)
                image = image.reshape(temp_fig.canvas.get_width_height()[::-1] + (4,))
                
                # Cerrar la figura temporal
                plt.close(temp_fig)
                
                # Mostrar la imagen renderizada en el subplot correspondiente
                ax[i, j].imshow(image)
                ax[i, j].set_title(plt.style.available[contador], fontsize=8)
                ax[i, j].axis('off')  # Quitar los ejes para mayor claridad
                
            contador += 1
        else:
            break

# Ajustar los espacios entre los gráficos
plt.tight_layout()
plt.show()
```
